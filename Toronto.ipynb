**Scraping the wikipedia page using Beautifulsoup**

```python
source=requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup=BeautifulSoup(source,'lxml')
print(soup.title)
from IPython.display import display_html
tab = str(soup.table)
display_html(tab,raw=True)
```

**Converting html page into pandas dataframe for cleaning and preprocessing**

```python
df_t=pd.read_html(tab)
df=df_t[0]
df.head()
```

**Cleaning and Preprocessing**

```python
df_na=df[df.Borough != 'Not assigned']
df_na
```

```python
df_na['Neighborhood']=np.where(df_na['Neighborhood']== 'Not assigned', df_na['Borough'], df_na['Neighborhood'])
df_na.reset_index()
```

```python
df_na.shape
```

**Importing the csv file of latitudes and longitudes**

```python
lat_long=pd.read_csv('http://cocl.us/Geospatial_data')
lat_long.head()
```

**Merging the preprocessed data and the latitudes and longitude data**

```python
df1=pd.merge(df_na,lat_long)
df1
```

**Separating the Boroughs of Toronto**

```python
Toronto_data=df1[df1['Borough'].str.contains('Toronto',regex=False)]
Toronto_data
```

**Obtaining the latitude and longitude of Toronto data**

```python
address = 'Toronto, Canada'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

**Mapping the Toronto data**

```python
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(Toronto_data['Latitude'], Toronto_data['Longitude'], Toronto_data['Borough'], Toronto_data['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```

**Clustering the Toronto data**

```python
k=5
toronto_clustering = Toronto_data.drop(['Postal Code','Borough','Neighborhood'],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(toronto_clustering)
kmeans.labels_
Toronto_data.insert(0, 'Cluster Labels', kmeans.labels_)
```

```python
Toronto_data
```

```python
map_clusters = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

# set color scheme for the clusters
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, neighborhood, cluster in zip(Toronto_data['Latitude'], Toronto_data['Longitude'], Toronto_data['Neighborhood'], Toronto_data['Cluster Labels']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
